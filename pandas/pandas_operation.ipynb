```python
import pandas as pd

# Sample dataset
data = {
    "Name": ["Ganesh", "Rahul", "Priya"],
    "Age": [21, 22, 20],
    "Course": ["BCA", "BBA", "BCA"]
}

df = pd.DataFrame(data)
print(df)
```

# 1.Only for csv file datasets
df = pd.read_csv("students.csv")
print(df)

# 2 .pd.read_table()
# read data from a delimited text fil

df = pd.read_table("students.txt", delimiter=",")
print(df)

## 3 . pd.read_excel()
# read only excel file
df = pd.read_excel("students.xlsx")
print(df)

# 4. pd.read_sql(query, connection_object)

👉 Reads data from a SQL database
import sqlite3

conn = sqlite3.connect("students.db")

df = pd.read_sql("SELECT * FROM students", conn)
print(df)

# 5. pd.read_json(json_string or file)

👉 Reads JSON data

df = pd.read_json("students.json")
print(df)

# 6. pd.read_html(url or file)

👉 Extracts tables from an HTML page

tables = pd.read_html("https://example.com")
df = tables[0]
print(df)

# 7. pd.DataFrame(data)

👉 Creates a DataFrame from dictionary, list, etc.

```python
data = {
    "Name": ["Ganesh", "Rahul"],
    "Marks": [85, 90]
}

df = pd.DataFrame(data)
print(df)
```

# 8. df.to_csv(filename)

👉 Writes DataFrame to CSV file

df.to_csv("output.csv", index=False)

# 9. df.to_excel(filename)

👉 Writes DataFrame to Excel file

df.to_excel("output.xlsx", index=False)

# 10. df.to_sql(table_name, connection_object)

👉 Writes data to SQL table

import sqlite3

conn = sqlite3.connect("students.db")

df.to_sql("students", conn, if_exists="replace", index=False)

# 11. df.to_json(filename)

👉 Writes data to JSON format

df.to_json("output.json", orient="records")
